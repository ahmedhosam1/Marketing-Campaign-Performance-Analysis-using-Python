```python
CampaignChannel_names = data.CampaignChannel.value_counts().index
CampaignChannel_names 
```

```python
CampaignChannel_count = list(data.CampaignChannel.value_counts())
CampaignChannel_count 
```

```python
plt.pie(CampaignChannel_count , labels = CampaignChannel_names, autopct= "%.2f%%")
plt.savefig("pie.jpg")
plt.show()
```

```python
plt.figure(figsize=(20,4))

plt.subplot(1,4,1)
plt.hist(data['Income'], bins=20)
plt.title('Income Distribution')

plt.subplot(1,4,2)
plt.hist(data['ConversionRate'], bins=20)
plt.title('Conversion Rate')

plt.subplot(1,4,3)
plt.scatter(
    data['Income'],
    data['ConversionRate'],
    alpha=0.3,
    s=10)
plt.title('Income vs Conversion')

plt.subplot(1,4,4)
plt.boxplot(data['Income'])
plt.title('Income Boxplot')

plt.tight_layout()
plt.show()
```
